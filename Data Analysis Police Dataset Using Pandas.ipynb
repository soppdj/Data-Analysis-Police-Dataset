## Data Cleaning

### Q1. To remove the column that contains missing values.

```python
# Checking value
police_data.count()
```

We have 65535 rows from dataset. If any column which does not have the 65535 rows it means missing value. To check it we will do in the next formula.

```python
# Checking missing value
police_data.isnull()
```

So, here from 65535 rows and 15 columns we found "True" missing value which are "country_name" and "search_type".

```python
# Checking total missing value
police_data.isnull().sum()
```

To find more detail, we can see the column that has "0" value it means the data is complate.  

```python
# Removing missing value
police_data.drop('country_name', axis =1)
```

Here, we have just removed "country_name" column from dataset. 

```python
# To verify deleting "country_name" column, we should use this formula to remove permanently.
police_data.drop('country_name', axis =1, inplace = True)
```

```python
# Removing missing value "search_type"
police_data.drop('search_type', axis =1)
```

```python
# To verify deleting "search_type" column, we should use this formula to remove permanently.
police_data.drop('search_type', axis =1, inplace = True)
```

```python
# Let's check again!
police_data
```

### Q2. For speeding, were men or women stopped more often?

```python
police_data.head(5)
```

```python
# Showing all violations "Speeding"
police_data[police_data.violation == 'Speeding']
```

```python
# Checking value_counts that shows each gender
police_data[police_data.violation == 'Speeding'].driver_gender.value_counts()
```

### Q3. Does gender effect who gets sarched during a stop?

```python
police_data.head(5)
```

```python
police_data.groupby('driver_gender').search_conducted.sum()
```

```python
police_data.search_conducted.value_counts()
```

### What is the men stop_duration?

```python
police_data.head(5)
```

```python
# Mapping and datatype casting
police_data.stop_duration.value_counts()
```

```python
# Changing value in column "stop_duration"
police_data['stop_duration'] = police_data['stop_duration'].map({'0-15 Min': 7.5, '16-30 Min': 23, '30+ Min': 45})
```

```python
# Checking one more time and make sure "stop_duration" column has replaced by mean!
police_data.head(5)
```

```python
police_data['stop_duration'].mean()
```
